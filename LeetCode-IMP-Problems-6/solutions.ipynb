```python
# Q1.Write a function that generates a specified number of independent and identically distributed
# (i.i.d.) draws from a uniform distribution between 0 and 1. The function should take the number of
# draws as an argument and return a list of random numbers.
# Note: Use a fixed random seed of 42 to ensure that the results are reproducible.
# Test Case #1
# Input: 5
# Output: &quot;[0.37, 0.95, 0.73, 0.6, 0.16]&quot;
# Description: This test case assesses the function&quot;s correctness in generating a small number of
# i.i.d. draws from the uniform distribution between 0 and 1.
# Test Case #2
# Input: 100
# Output: &quot;A list of 100 random numbers between 0 and 1, e.g., [0.134, 0.245, 0.756, ..., 0.890]. &quot;
# Description: This test validates the function&quot;s ability to handle a larger number of i.i.d. draws,
# ensuring the output remains within the expected range and maintains the properties of the
# distribution.
# Test Case #3
# Input: 0
# Output: &quot;An empty list, [].&quot;
# Description: This case challenges the function&quot;s handling of the edge case where no draws are
# requested, ensuring it returns an empty list appropriately.
```

```python
import numpy as np

def Generate_Random_Number(n):
    np.random.seed(32)
    d=np.random.uniform(0,1,n)
    return d

Generate_Random_Number(10)
```

```python
# Write a function to reverse a string.
# Constraints
#  The input variable string must be of type str.
#  There are no specific constraints on the length or content of the string.
# Test Case #1
# Input: &quot;G3n3r@tiv3AI&quot;
# Output: &quot;IA3v1t@r3n3G&quot;
# Description: This test case examines the function’s capacity to handle a mix of uppercase
# letters, lowercase letters, numbers, and special characters without altering their order other than
# reversing the string.
# Test Case #2
# Input: &quot;Le3tsWriT#S0meTestCases!&quot;
# Output: &quot;!sesaCtseT0Sem#TirWsTe3L&quot;
# Description: This case tests the function with a longer string composed of varied characters,
# including uppercase and lowercase letters, numbers, and special characters, to ensure the
# reversal is accurately managed.
```

```python
def Reversing_String(s):
    rev=""
    for i in range(1,len(s)+1):
        rev+=s[-i]
    return rev

Reversing_String("sambhaji")
```

```python
# Q3. Given an array list of non-negative integers, return the maximum product of two numbers
# possible.
# Input: arr[] = [1, 4, 3, 6, 7, 0]
# Output: 42
# Explanation: 6 and 7 have the maximum product.
```

```python
arr=[1, 4, 3, 6, 7, 0]

def max_product(arr):
    max_pro=float('-inf')
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if (arr[i]*arr[j])>max_pro:
                max_pro=(arr[i]*arr[j])
    return max_pro

max_product(arr)
```

```python
# Q4. Bheem has N friends and he has a ladoo for each day given an list denoting on which
# day i th index friend wants a ladoo if bheem is unable to give a ladoo to the friend he loses
# his friendship with them find out the maximum number of friends he can have at the end
# Example: N=5
# List of days=[3,3,1,2,4]
# Output=4
```

```python
m=[3,3,1,2,4]

def max_friends(m):
    new_m=[]
    for i in m:
        if i not in new_m:
            new_m.append(i)
    return len(new_m)
    
max_friends(m)
```

```python
# Q5 Write a function that samples the flip of a biased coin for N times.
# Test Case #1
# Input: {&quot;N&quot;: 10, &quot;probability&quot;: 0.5}
# Output: [1, 1, 0, 0, 1, 0, 1, 0, 0, 1]
# Description: This test case checks the function with a fair coin flip (probability 0.5) for 10 trials.
# Test Case #2
# Input: {&quot;N&quot;: 5, &quot;probability&quot;: 0.8}
# Output: [1, 1, 1, 1, 1]
# Description: This test case uses a biased coin with a higher probability of heads (0.8) for 5
# trials.
# Test Case #3
# Input: {&quot;N&quot;: 15, &quot;probability&quot;: 0.1}
```

```python
import numpy as np
n=10
p=0.5
def flip_coin(n,p):
    if p==0.5:
        num=[0 if i<(n//2) else 1 for i in range(n)]
        np.random.shuffle(num)
        return num
    elif p>0.5:
        return [1 for i in range(n)]
    elif p<0.5:
        return [0 for i in range(n)]
    else:
        return "Wrong P"

flip_coin(n,p)
```
