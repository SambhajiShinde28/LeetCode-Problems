```python
# Q3. Count the number of user events performed by MacBookPro users. Output the result
# along with the event name. Sort the result based on the event count in the descending
# order.
# Table:Â playbook_events
# user_id:bigint
# occurred_at:timestamp
# event_type:text
# event_name:text
# location:text
# device:text
```

```python
SELECT 
    event_name, 
    COUNT(*) AS event_count
FROM 
    playbook_events
WHERE 
    device = 'MacBookPro'
GROUP BY 
    event_name
ORDER BY 
    event_count DESC;
```

```python
# Q4. Calculate the total revenue from each customer in March 2019. Include only
# customers who were active in March 2019.
# Output the revenue along with the customer id and sort the results based on the revenue
# in descending order.
# Table:orders
# id:bigint
# cust_id:bigint

# order_date:date
# order_details:text
# total_order_cost:bigint
```

```python
SELECT 
    cust_id,
    SUM(total_order_cost) AS total_revenue
FROM 
    orders
WHERE 
    order_date BETWEEN '2019-03-01' AND '2019-03-31'
GROUP BY 
    cust_id
ORDER BY 
    total_revenue DESC;
```

```python
# Q5. Find the employee with the highest salary per department. Output the department
# name, employee's first name along with the corresponding salary.
# Table:Employee:
# id:bigint
# first_name:text
# last_name:text
# age:bigint
# sex:text
# employee_title:text
# department:text
# salary:bigint
# target:bigint
# bonus:bigint
# email:text
# city:text
# address:text
# manager_id:bigint
```

```python
select 
    first_name,
    department,
    salary
from 
    employee
order by 
    department
having 
    max(salary)
```
