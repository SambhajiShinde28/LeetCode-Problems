```python
#Q3. 10*10 array and finding the minimum and max element
import numpy as np

a=[]
def ten_array():
    new_a=[]
    for i in range(10):
        a.append(list(np.random.randint(10,100,size=10)))
    for i in a:
        m=min(i)
        ma=max(i)
        new_a.append(m)
        new_a.append(ma)
    print("Original Array : ",a)
    return min(new_a),max(new_a)

ans=ten_array()
print("Minimum : ",ans[0])
print("Maximum : ",ans[1])
```
