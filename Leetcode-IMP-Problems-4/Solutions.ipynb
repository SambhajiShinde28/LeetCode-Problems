```python
# Q1.Given an array of integers nums and an integer target, return indices
# of the two numbers such that they add up to target.
# You may assume that each input would have exactly one solution, and
# you may not use the same element twice.
# You can return the answer in any order.
#  
# Example 1:
# Input: nums = [2,7,11,15], target = 9
# Output: [0,1]
# Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
# Example 2:
# Input: nums = [3,2,4], target = 6
# Output: [1,2]
# Example 3:
# Input: nums = [3,3], target = 6
# Output: [0,1]
```

```python
n=[3,3]
target=6
def sum_elements(n,target):
    for i in range(len(n)):
        for j in range(i+1,len(n)):
            if n[i]+n[j]==target and i!=j:
                return [i,j]
sum_elements(n,target)
```

```python
n=[2,7,11,15]
target=9
def sum_elements(n,target):
    for i in range(len(n)):
        d=target-n[i]
        for j in range(i+1,len(n)):
            if d==n[j]:
                return [i,j]
sum_elements(n,target)
```

```python
n=[2,7,11,15]
target=9
def sum_elements(n,target,summ,left,right):
    summ=n[left]+n[right]
    
sum_elements(n,target,summ=0,left=0,right=1)
```

```python
# Q2.Find the common elements between two lists using dictionary
```

```python
a=[7,4,8,3,9,2,9,3]
b=[2,1,3,4,5,6,7,6,5,4,3,2]

def common_element(a,b):
    a=list(set(a))
    b=list(set(b))
    d={}
    print("list a : ",a)
    print("list b : ",b)
    for i in a:
        if i in b:
            d[i]=1
    return list(d.keys())

common_element(a,b)
```

```python
# Q3.Write Your own insert() and Search() in List without using any predefined
# function
```

```python
n=[4,5,6,7,8,9,8,7,6,5]
pos=20
new_element=99

def own_insert(n=list,new_element=int,position=int)->list:
    if position>len(n):
        return "index out of bound"
    else:
        n[position-1]=new_element
        return n

own_insert(n,new_element,pos)
```

```python
n=[4,5,6,7,8,9,8,7,6,5]
se=6

def own_search(n,se):
    for i in range(len(n)):
        if n[i]==se:
            return i
    else:
        return "Element not found"
own_search(n,se)
```

```python
# Q5.Write a Binary search Algorithm recursive code
```

```python
n=[9,9,8,7,2,3,4,5,8,7,6,7,8]
n=sorted(n)
start=0
end=len(n)-1
se=7

print("This is n : ",n)

def binary_search(n,start,end,se):
    mid=(start+end)//2
    if n[mid]>se:
        end-=1
    elif n[mid]<se:
        start+=1
    elif n[mid]==se:
        return mid
    elif start>end:
        return "element not found"

    return binary_search(n,start,end,se)

binary_search(n,start,end,se)
```

```python
n=[9,9,8,7,2,3,4,5,8,7,6,7,8]
n=sorted(n)
start=0
end=len(n)-1
se=7

print("This is n : ",n)

def binary_search(n,start,end,se):
    mid=(start+end)//2
    if n[mid]>se:
        return binary_search(n,start,mid-1,se)
    elif n[mid]<se:
        return binary_search(n,mid+1,end,se)
    elif n[mid]==se:
        return mid
    elif start>end:
        return "element not found"

binary_search(n,start,end,se)
```

```python
# Q6.Merge two sorted List using iterative approach
```

```python
list1=[5,6,8,8,9]
list2=[1,2,4,5,7]

def sort_list(l1,l2):
    l=list1+list2

    for i in range(len(l)):
        for j in range(i+1,len(l)):
            if l[j]<l[i]:
                temp=l[i]
                l[i]=l[j]
                l[j]=temp
    
    return l

sort_list(list1,list2)
```
